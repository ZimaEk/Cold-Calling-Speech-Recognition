# Rubert

This is a DeepPavlov/rubert-base-cased-conversational model trained on aggregated corpus of 351.797 texts.

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from transformers import pipeline

model_name = "blanchefort/rubert-base-cased-sentiment"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

classifier = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
```

```python
# Открытие файла без разделения на строки
with open('audio_5.txt', 'r', encoding='utf-8') as file:
    audio_5_full = file.readlines()
classifier(audio_5_full)
```

```python
# Выделяем реплики МОПа (для сокращения токенов)
#audio_5_MOP = [line.strip() for line in audio_5 if line.startswith("[SPEAKER_00]")]
#classifier(audio_5_MOP)

# Выделяем реплики РОПа (для сокращения токенов)
#audio_5_ROP = [line.strip() for line in audio_5 if line.startswith("[SPEAKER_01]")]
#classifier(audio_5_ROP)
```

```python
#Анализ тональности для каждого текста
results = []
for text in audio_5:
    # Получаем предсказание (указываем truncation=True для длинных текстов)
    result = classifier(text, truncation=True)[0]
    results.append({
        "text": text,
        "sentiment": result["label"],
        "confidence": result["score"]  # Дополнительно: уверенность модели
    })
```

```python
# Создаем DataFrame
df_polarity = pd.DataFrame(results)
df_polarity
```

```python
# Функция для разделения текста на спикера и содержание
import re
def split_speaker(text):
    match = re.match(r'^(\[SPEAKER_\d{2}\])(.*)', text)
    if match:
        return match.group(1), match.group(2).strip()
    return None, text

# Применяем функцию и создаем новые столбцы
df_polarity[['speaker', 'text']] = df_polarity['text'].apply(lambda x: pd.Series(split_speaker(x)))

# Удаляем исходный столбец (если нужно)
# df = df.drop(columns=['text'])
```

```python
df_polarity
```

```python
# Добавим номер реплики и подпись спикера
df_polarity['replica_num'] = range(1, len(df) + 1)
df_polarity['replica_label'] = df_polarity.apply(lambda row: f"{row['replica_num']}\n({row['speaker']})", axis=1)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Визуализация динамики тональности
plt.figure(figsize=(10, 5))

# Цвета для тональностей
colors = {'POSITIVE': '#4CAF50', 'NEGATIVE': '#F44336', 'NEUTRAL': '#2196F3'}

# Рисуем линии и точки для каждой реплики
for i, row in df.iterrows():
    plt.scatter(
        row['replica_num'],
        row['confidence'],
        color=colors[row['sentiment']],
        s=100,
        label=row['sentiment'] if i == 0 else ""
    )
    plt.text(
        row['replica_num'],
        row['confidence'] + 0.03,
        row['sentiment'],
        ha='center',
        fontsize=10
    )

# Соединяем точки линиями для наглядности
plt.plot(
    df_polarity['replica_num'],
    df_polarity['confidence'],
    linestyle='--',
    color='gray',
    alpha=0.5
)

# Настраиваем ось X с подписями спикеров
plt.xticks(
    df_polarity['replica_num'],
    df_polarity['replica_label'],
    rotation=45,
    ha='right'
)

plt.title('Динамика тональности в диалоге')
plt.xlabel('Номер реплики')
plt.ylabel('Вероятность тональности')
plt.ylim(0, 1.1)
plt.xticks(df['replica_num'])
plt.grid(True)

# Увеличиваем расстояние между репликами
plt.xlim(0.5, len(df) + 0.5)  # Добавляем отступы по краям

# Убираем дубликаты в легенде
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys(), title='Тональность')

plt.show()

# Преобладающая тональность (можно добавить анализ по спикерам)
average_prob = df_polarity.groupby('sentiment')['confidence'].mean()
dominant_sentiment = average_prob.idxmax()

#print(f"\nПреобладающая тональность в диалоге: {dominant_sentiment}")
print("Средние вероятности по тональностям:")
print(average_prob)

# Дополнительно: анализ тональности по спикерам
speaker_sentiment = df_polarity.groupby('speaker')['confidence'].mean()
print("\nСредняя вероятность тональности по спикерам:")
print(speaker_sentiment)
```

```python
# 1. Анализ преобладающей тональности (по количеству реплик)
sentiment_counts = df['sentiment'].value_counts()
dominant_sentiment = sentiment_counts.idxmax()

# Процентное соотношение
total_replicas = len(df)
sentiment_percent = (sentiment_counts / total_replicas) * 100

print("═" * 50)
print("📊 Анализ тональности по количеству реплик:")
print(f"• Преобладающая тональность: {dominant_sentiment}")
print("• Распределение:")
for sentiment, percent in sentiment_percent.items():
    print(f"  {sentiment}: {percent:.1f}% ({sentiment_counts[sentiment]} из {total_replicas} реплик)")
print("═" * 50)
```

# NER with NATASHA

Персона (PER), локация (LOC), организация (ORG)

```python
# Открытие файла в режиме чтения ('r')
with open('audio_8.txt', 'r', encoding='utf-8') as file:
    audio_8 = file.readlines()
    print(audio_8)
```

```python
audio_8_cln = []
for sent in audio_8:
  #one = sent.replace('[SPEAKER_00]', "КЛИЕНТ:")
  #two = one.replace('[SPEAKER_01]', "Менеджер:")
  sent = sent.replace('\n', "")
  audio_8_cln.append(sent)
```

```python
audio_8_cln
```

```python
# Установка необходимых пакетов
from natasha import (
    Segmenter,
    MorphVocab,

    NewsEmbedding,
    NewsMorphTagger,
    NewsSyntaxParser,
    NewsNERTagger,

    PER,
    NamesExtractor,

    Doc
)
```

```python
#Инициализация Natasha
segmenter = Segmenter()
emb = NewsEmbedding()
morph_tagger = NewsMorphTagger(emb)
ner_tagger = NewsNERTagger(emb)

# Перебираем предложения
results = []

# Разбиваем текст на предложения и обрабатываем
for sent in audio_8_cln:
    doc = Doc(sent)
    doc.segment(segmenter)
    doc.tag_ner(ner_tagger)

    # Если в предложении есть NER, добавляем в результаты
    for span in doc.spans:
        results.append({
            "text": sent,
            "type": span.type,
            "ner": span.text
        })
```

```python
#Создаем DataFrame
import pandas as pd
df_ner = pd.DataFrame(results)
df_ner

# Сохраняем в CSV (опционально)
#df.to_csv("ner_results.csv", index=False, encoding="utf-8")
```
